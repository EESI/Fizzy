```python
import json 
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

mapf = pd.read_csv("american-gut-mf.txt", sep="\t")[["#SampleID","SEX"]]
dobj = json.loads(open("ag-output.biom","U").read())
feats =  [d["metadata"]["taxonomy"] for d in dobj["rows"]]
otu_ids =  [d["id"] for d in dobj["rows"]]
sample_ids = np.array([np.float(d["id"]) for d in dobj["columns"]])
labels = []
for n in range(len(sample_ids)):
    q = np.where(sample_ids[n] == mapf["#SampleID"])[0]
    if len(q) > 0:
        labels.append(mapf["SEX"][q].tolist()[0])
labels = np.array(labels)
male_data= np.array(dobj["data"])[np.where(labels == "male")]
female_data = np.array(dobj["data"])[np.where(labels == "female")]

def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)


fig = plt.figure(dpi=500)    
# not sure why these plots are throwing an error
try:
    bpl = plt.boxplot(male_data[:,:10], positions=np.array(xrange(len(male_data[:,:10])))*2.0-0.4, sym='', widths=0.6)
except:
    bpl = plt.gca()
    print "Caught error."
    
try:
    bpr = plt.boxplot(female_data[:,:10], positions=np.array(xrange(len(female_data)))*2.0+0.4, sym='', widths=0.6)
except:
    bpr = plt.gca()
    print "Caught error."

plt.ylim([0,.005])
plt.xlim([-1,19])
ax1 = plt.gca()

for n,f in enumerate(feats[:10]):
    fnew = f.replace("[","").replace("]","").replace('"',"").replace(" ","")
    print "("+str(2*n)+") "+fnew+" (GGID:"+str(otu_ids[n])+")"

plt.title("Rel. Abundance of Top OTUs")
plt.ylabel('rel. abundance', fontsize=20)
plt.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica'],'size':18})
plt.tight_layout()

plt.savefig("ag-mf-relabunbar.pdf")
```
