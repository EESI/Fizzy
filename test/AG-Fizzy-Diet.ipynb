# About

Run NPFS on data collected from the American Gut Project. We isloated the gut samples then created a dataset from the remaining samples that were either from: (a) omnivores, or (b) vegans (i.e., vegetarians are NOT included). We run NPFS with `MIM` then plot the relative abundances of the features that were selected for the omnivores and vegans, respectively. 

# Run NPFS

```python
%%bash
npfs -l DIET_TYPE -f MIM -i /Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Diet-OmniVegan.biom -m /Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Diet-OmniVegan.txt -o tmp.txt -n 100 -c 5 -b 500 -s -r ../files/ag-diet-result.biom
cat tmp.txt | sed -e "s/[a-z]\_\_//g" -e "s/,,//g" -e 's/,\t/\t/g' > ../files/results-npfs.txt
rm tmp.txt
```

# Plot the Results

```python
import json 
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

mapf = pd.read_csv("/Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Diet-OV.txt", sep="\t")
dobj = json.loads(open("/Users/gditzler/Git/Miscellaneous-IPython-Notebooks/files/ag-diet-result.biom","U").read())
feats =  [d["metadata"]["taxonomy"] for d in dobj["rows"]]
sample_ids = np.array([d["id"] for d in dobj["columns"]])
labels = []
for n in range(len(sample_ids)):
    q = np.where(sample_ids[n] == mapf["#SampleID"])[0]
    if len(q) > 0:
        labels.append(mapf["DIET_TYPE"][q].tolist()[0])
labels = np.array(labels)
meat_data= np.array(dobj["data"])[np.where(labels == "Omnivore")]
vegy_data = np.array(dobj["data"])[np.where(labels == "Veggie")]

fig = plt.figure(dpi=500)    
# not sure why these plots are throwing an error
try:
    bpl = plt.boxplot(meat_data[:,:10], positions=np.array(xrange(len(meat_data[:,:10])))*2.0-0.4, sym='', widths=0.6)
except:
    bpl = plt.gca()
    print "Caught error."
    
try:
    bpr = plt.boxplot(vegy_data[:,:10], positions=np.array(xrange(len(vegy_data)))*2.0+0.4, sym='', widths=0.6)
except:
    bpr = plt.gca()
    print "Caught error."

plt.ylim([0,.005])
plt.xlim([-1,19])
ax1 = plt.gca()

plt.xticks(np.arange(0, 19+1, 2.0)) 
delta=.0005


for n,f in enumerate(feats[:10]):
    fnew = f.replace("[","").replace("]","").replace('"',"").replace(" ","")
    plt.text(21, .0045-n*delta, "("+str(2*n)+") "+fnew,fontsize=19)

plt.title("Rel. Abundance of Top OTUs")
plt.ylabel('rel. abundance', fontsize=20)
plt.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica'],'size':20})
#plt.rc('figure', **{'autolayout': True})
```
